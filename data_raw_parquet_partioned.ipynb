```python
year = 2020

for month in range(1, 13):
    print(f'Processing data for {year}/{month:02d}')

    # Input and output paths
    input_path = f'data/raw/green/{year}/{month:02d}/green_tripdata_{year}_{month:02d}.parquet'
    output_path = f'data/pq/green/{year}/{month:02d}/'

    # Read the Parquet file (schema is automatically inferred)
    df_green = spark.read.parquet(input_path)

    # Optional: Print schema for verification (remove in production)
    print(f"Schema for {year}/{month:02d}:")
    df_green.printSchema()

    # Partition and write back as Parquet
    df_green \
        .repartition(4) \
        .write \
        .mode('overwrite') \
        .parquet(output_path)
```
